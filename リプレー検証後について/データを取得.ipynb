```python
url='https://npb.jp/bis/2021/stats/bat_c.html'
dfs=pd.read_html(url)
#不要な一行目を削除
df=dfs[0]
df=df.drop(0) 
```

```python
display(df)
```

```python
#column名を付け直す
df.columns=['rank','name','team','ba','g','pa','ab','r','h','_2b','_3b', 'hr', 'tb', 'rbi', 'sb', 'cs', 'sh', 'sf', 'bb','bb(intention)', 'hbp', 'so', 'dp','slg', 'obp']

#1行目を削除(カラム名と重複しているため)
#df = df.drop(1)

#インデックスを振り直す
df.index = np.arange(1,len(df)+1)
display(df)
```

```python
#文字列をfloat型に変換
import numpy as np
#df['year'] = df['year'].fillna(0).astype(np.float64)
df['g'] = df['g'].fillna(0).astype(np.float64)
df['pa'] = df['pa'].fillna(0).astype(np.float64)
df['ab'] = df['ab'].fillna(0).astype(np.float64)
df['r'] = df['r'].fillna(0).astype(np.float64)
df['h'] = df['h'].fillna(0).astype(np.float64)
df['_2b'] = df['_2b'].fillna(0).astype(np.float64)
df['_3b'] = df['_3b'].fillna(0).astype(np.float64)
df['hr'] = df['hr'].fillna(0).astype(np.float64)
df['tb'] = df['tb'].fillna(0).astype(np.float64)
df['rbi'] = df['rbi'].fillna(0).astype(np.float64)
df['sb'] = df['sb'].fillna(0).astype(np.float64)
df['cs'] = df['cs'].fillna(0).astype(np.float64)
df['sh'] = df['tb'].fillna(0).astype(np.float64)
df['sf'] = df['sf'].fillna(0).astype(np.float64)
df['bb'] = df['bb'].fillna(0).astype(np.float64)
df['hbp'] = df['hbp'].fillna(0).astype(np.float64)
df['so'] = df['so'].fillna(0).astype(np.float64)
df['dp'] = df['dp'].fillna(0).astype(np.float64)
df['ba'] = df['ba'].fillna(0).astype(np.float64)
df['slg'] = df['slg'].fillna(0).astype(np.float64)
df['obp'] = df['obp'].fillna(0).astype(np.float64)
```

```python
#打撃指標を計算してデータフレームに追加
#df['obp']=(df['h']+df['bb']+df['hbp'])/(df['ab']+df['bb']+df['hbp']+df['sf'])
df['ops']=df['obp']+df['slg']
df['IsoP']=df['slg']-df['ba']
df['wOBA']=(0.7*(df['bb']+df['hbp'])+(0.9*df['h'])+(1.3*df['_2b'])+(1.6*df['_3b'])+(2.0*df['hr']))/(df['ab']+df['bb']+df['hbp']+df['sf'])
display(df)
```

```python
#グラフで描画
import seaborn as sns
sns.pointplot(x="rank",y="IsoP",data=df)
```

```python
### 相関を見る
print(df.corr())
```

```python
#規定打席到達者数の平均打率
np.mean(df['ba'])
```
